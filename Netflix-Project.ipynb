### 8) What is the maximum duration of a Movie/Show on Netflix ?

```python
netflix['Duration'].unique()
```

```python
netflix['Duration'] = netflix['Duration'].str.replace(' min','')
netflix['Duration'] = netflix['Duration'].str.replace(' Seasons','')
netflix['Duration'] = netflix['Duration'].str.replace(' Season','')
```

```python
netflix['Duration']= netflix['Duration'].astype(np.int64)
```

```python
netflix[netflix['Duration']==netflix['Duration'].max()]
#max duration is 312 min
```
